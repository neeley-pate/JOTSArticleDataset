```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame with a "Date" column
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/CME/JOTS Revisions/English Used Tweets.csv")
df = pd.DataFrame(data)
# print(df["Date"])
df["Date"] = pd.to_datetime(df["Date"]).dt.to_period('M')  # Convert the "Date" column to datetime format

# print(df["Date"])
# Create a frequency plot
plt.figure(figsize=(10, 6))
df["Date"].value_counts().sort_index().plot()
plt.xlabel("Date")
plt.ylabel("Frequency")
plt.title("Frequency of Relevant English Tweets per Month")
# plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
# plt.tight_layout()
plt.show()
```
