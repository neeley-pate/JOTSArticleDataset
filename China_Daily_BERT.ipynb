```python
import pandas as pd

# Define list of headlines
# there are in fact no exact duplicates
all_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CME/China Daily.csv')
headlines = all_data['Headline']

headline_times = all_data['Date'].values.tolist()

# Tokenize headlines
tokenized_headlines = []
for headline in headlines:
  tokenized_headlines.append(headline.lower().split())
processed_tokenized_headlines = []
print(len(tokenized_headlines))
for headline in tokenized_headlines:
  if 'Xinjiang' in headline or 'China' in headline or 'Uyghur' in headline or 'xinjiang' in headline or 'china' in headline or 'uyghur' in headline or 'uygur' in headline or 'Uygur' in headline:
    if 'COVID' in headline or 'covid'  in headline or 'Covid' in headline or 'virus' in headline:
      if 'xinjiang' in headline or 'uyghur' in headline or 'uygur' in headline:
        processed_tokenized_headlines.append(headline)
    else:
      processed_tokenized_headlines.append(headline)

with open('/content/drive/MyDrive/Colab Notebooks/CME/english_stopwords.txt', encoding="utf8") as f:
    stopwords = [w.strip() for w in f.readlines()]

tokenized_headlines = processed_tokenized_headlines
processed_tokenized_headlines = []
for headline in tokenized_headlines:
  temp_str = []
  for word in headline:
    if (word.lower() not in stopwords) and (len(word) > 1):
      # if ('http' not in word) and ('@' not in word) and ('#' not in word):
        if ('xinjiang' not in word) and ('china' not in word) and ('Xinjiang' not in word) and ('China' not in word):
          if ('Uyghur' not in word) and ('Uygur' not in word) and ('uyghur' not in word) and ('uygur' not in word):
            is_english = True
            for char in word:
              if ord(char) > 255:
                is_english = False
              if char.lower() not in ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']:
                word = word.replace(char, '')
            if is_english:
              temp_str.append(word)
  if len(temp_str) > 0:
    processed_tokenized_headlines.append(temp_str)

# tokenized_headlines = processed_tokenized_headlines
print("English headlines: ", len(processed_tokenized_headlines))

import gensim
from gensim import corpora
import pandas as pd
!pip install bertopic
# from bertopic import BERTopic
from umap import UMAP

# Create dictionary of all words
dictionary = corpora.Dictionary(processed_tokenized_headlines)

# Convert tokenized headlines to bag-of-words vectors
corpus = [dictionary.doc2bow(headline) for headline in processed_tokenized_headlines]

# corpus_seed = [dictionary_seed.doc2bow(headline) for headline in seed_topics]
headlines_for_model = [' '.join(str(x) for x in headline) for headline in processed_tokenized_headlines]

headlines_for_model = headlines_for_model

umap_model = UMAP(n_neighbors=12,
                  n_components=5,
            min_dist=0.0,
            metric='cosine',
            low_memory=False,
          random_state=123)

# Further reduce topics
# print(len(headlines_for_model))
topic_model = BERTopic(umap_model=umap_model)
# topic_model = BERTopic()

topic_model.fit_transform(headlines_for_model)

topic_model.reduce_topics(headlines_for_model, nr_topics=6)

topics, probs = topic_model.fit_transform(headlines_for_model)


topic_counts = pd.Series(topics).value_counts()

print(topic_counts[:50])
# fig = topic_model.visualize_topics()
# fig.show()

fig3 = topic_model.visualize_documents(headlines_for_model)
fig3.show()

from gensim.models.coherencemodel import CoherenceModel

# from gensim.models.coherencemodel import CoherenceModel
# # Extract the top N words for each topic
dictionary_of_topics = topic_model.topic_representations_
top_words = []
for i in range(len(topic_counts) - 1):
  new_topic = []
  for items in dictionary_of_topics[i]:
    print(i)
    print(items[0])
    new_topic.append(items[0])
  top_words.append(new_topic)
  print("\n")
# top_words = [model.topic_representations_[i] for i in range(4)]
# print(top_words)
# Calculate the coherence score using UMass
cm_umass = CoherenceModel(topics=top_words, corpus=corpus, dictionary=dictionary, coherence='u_mass')
coherence_umass = cm_umass.get_coherence()
print(coherence_umass)

# # Access the UMAP parameters
# n_neighbors = umap_model.n_neighbors
# n_components = umap_model.n_components
# metric = umap_model.metric

# print("UMAP Parameter Values:")
# print("Number of Neighbors:", n_neighbors)
# print("Number of Components:", n_components)
# print("Metric:", metric)
```
